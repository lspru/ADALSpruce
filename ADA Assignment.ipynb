```python
import pandas as pd
import numpy as np
```

```python
import json

data_file = 'arxiv.json'

def get_metadata():
    with open(data_file, 'r') as f:
        for line in f:
            yield line
```

```python
i_d = []
titles = []
comments = []
journal_ref = []
categories = []
abstracts = []

metadata = get_metadata()
for paper in metadata:
    paper_dict = json.loads(paper)
    ref = paper_dict.get('journal-ref')
    try:
        i_d.append(paper_dict.get('id'))
        titles.append(paper_dict.get('title'))
        comments.append(paper_dict.get('comments'))
        journal_ref.append(paper_dict.get('journal-ref'))
        categories.append(paper_dict.get('categories'))
        abstracts.append(paper_dict.get('abstracts'))
    except:
        pass
```

```python
df = pd.DataFrame({
        'id' : i_d,
        'title': titles,
        'comments' : comments,
        'journal-ref' : journal_ref,
        'categories' : categories,
        'abstract': abstracts
        })
df.head()
```

```python
from collections import Counter
from nltk.corpus import stopwords
import string
import nltk
import re
from string import punctuation
nltk.download('stopwords')

stop = set().union(stopwords.words('english'))
```

```python
process = df
process
```

```python
process.drop(columns=['processed title'])
```

```python
def preprocess_text(text):
    text = text.lower()
    text = re.sub(f"[{re.escape(punctuation)}]", "", text)
    text = " ".join(text.split())
    return text

process['processed title'] = process['title'].map(preprocess_text)
```

```python
process['processed title'] = process['processed title'].apply(lambda x: ' '.join\
                                                              ([word for word in x.split() if word not in (stop)]))
```

```python
process
```

```python
def popular_words(series):
    process['processed title'] = series.apply(lambda x:str(x).split())
    top = Counter([item for sublist in process['processed title'] for item in sublist])
    temp = pd.DataFrame(top.most_common(10))
    temp.columns = ['Common_words','count']
    return temp

popular_words(process['processed title'])
```
